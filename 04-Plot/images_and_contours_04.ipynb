```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150


from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(np.random.normal(size=(512,512)), sigma=10)
```

```python
# Basico
plt.imshow(rands2d)
plt.colorbar()
```

```python
# Limites del grafico
plt.imshow(rands2d, extent=[-200,100,-100,100])
plt.colorbar()
```

```python
# Minimo y maximo valor de la escala
plt.imshow(rands2d, vmax=0.1, vmin= -0.05) 
plt.colorbar()
```

```python
# Aspect ratio
plt.imshow(rands2d, aspect=0.5625) # 0.... 1 horizontal, vertical
```

```python
# Interpolacion
small = np.arange(0,9,1).reshape(3,3)
print(small)
plt.imshow(small, interpolation='bilinear')

plt.matshow(small)  # helper function, deshabilita la interpolacion y usa los valores en el centro

"""
[None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
           'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
           'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']
"""
```

```python
# CONTORNOS

# Basico
srands2d = gaussian_filter(rands2d, sigma=100)
plt.imshow(srands2d)
plt.colorbar()
```

```python
# Contorno
plt.contour(srands2d)
plt.colorbar()
```

```python
# Niveles (3D)
plt.contour(srands2d, levels=[0,0.003])
```

```python
# Numero de contornos
plt.contour(srands2d, 30)
```

```python
# Colores del contorno
plt.contour(srands2d, 4, colors=['r','b'])
```

```python
# Contornos llenos
plt.contourf(srands2d, 10)
```

```python
# Rejillas
plt.contourf(srands2d, 3, colors='none', hatches=['*','-'])
```

```python
# Estilo de la linea
plt.contour(srands2d, linestyles=[':','--','-'])
```

```python
# labels
contours = plt.contour(srands2d, 15)
plt.clabel(contours)
```

```python
# Multiple
contours = plt.contour(srands2d, levels,colors=['k']) # levels=[0],
plt.imshow(srands2d)
plt.clabel(contours)
```

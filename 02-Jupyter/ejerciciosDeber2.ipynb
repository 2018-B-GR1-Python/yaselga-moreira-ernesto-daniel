```python
import numpy as np
from scipy import misc
from Crypto.Cipher import DES
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
# los bloques deben ser de 8 caracteres (se rellena con espacios blancos).
mensaje = "Keep it secret, keep it safe    "
print("Plain text:")
print(mensaje)

# clave DES
cipher = DES.new('12345678')
# crifado
cipher_text = cipher.encrypt(mensaje)

# cipher text
print("Cipher text: ")
print(cipher_text)

# descifrado, quito espacios con strip()
cipher = DES.new('12345678')
plain_text = cipher.decrypt(mensaje).strip()
print("Plain text:")
print(mensaje)
```

```python
import numpy as np
# pip install scipy
from scipy import misc
```

```python
mapache = misc.face()  # Cara de un mapache
```

```python
mapache.shape
```

```python
type(mapache)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(mapache)
plt.show()
```

```python
mapache_parcial = mapache[384:,512:,:]
plt.imshow(mapache_parcial)
plt.show()
```

```python
mapache_parte_uno, mapache_parte_dos = np.split(mapache,2,axis=1)
plt.imshow(mapache_parte_uno)
plt.show()
plt.imshow(mapache_parte_dos)
plt.show()
```

```python
mapache_frankenstein = np.concatenate(
    (mapache_parte_uno,mapache_parte_dos), axis=1
)
plt.imshow(mapache_frankenstein)
plt.show()
```

```python
import numpy as np
```

```python
salarios = np.genfromtxt('data/salary.csv', delimiter=',')
print(salarios)
```

```python
salarios.max() # muestra el mayor salario
```

```python
np.argmax(salarios) # muestra el índice del mayor salario
```

```python
np.argmin(salarios) # muestra el indice del salario más bajo
salarios[282]
```

```python
np.argsort(salarios) # odenados de menir a mayor los índices
```

```python
np.argsort(salarios, kind = 'mergesort') # odenados de menir a mayor los índices
```

```python
indices_mayores_a_50_mil = np.where(salarios>50000)
indices_mayores_a_50_mil
salarios[indices_mayores_a_50_mil]
```

```python
condicion_media = salarios > np.mean(salarios)
salarios[condicion_media]
```

```python
np.extract(condicion_media, salarios)
```
